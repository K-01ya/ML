```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('/content/drive/MyDrive/cmc/prac/prac3_flask/kc_house_data.csv')
```

```python
data.head(5)
```

```python
X = data.drop(columns=['id', 'price'])
y = data['price']
```

```python
X['date'] = X['date'].apply(pd.to_datetime)
X['date'] = (X['date'] - X['date'].min()).dt.days
```

```python
data.dtypes
```

```python
X.shape
```

```python
y = y.to_numpy()
X = X.to_numpy()
```

```python
from RandomForest_GradientBoost import RandomForestMSE, GradientBoostingMSE
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
rf = RandomForestMSE(n_estimators=10, max_depth=5, feature_subsample_size=0.5)
```

```python
rf.fit(X_train, y_train, X_val=X_test, y_val=y_test)
```

```python
gb = GradientBoostingMSE(n_estimators=100, learning_rate=0.3, max_depth=6, feature_subsample_size=0.7)
```

```python
gb.fit(X_train, y_train, X_val=X_test, y_val=y_test)
```

##Эксперименты

```python
from sklearn.model_selection import GridSearchCV
import time


parametrs = {'max_depth': [2, 4, 6, 8, 10, 12, None], 'n_estimators': [1, 20, 50, 100, 150, 200], 'feature_subsample_size': [0.1, 0.3, 0.5, 0.7, 1]}

rf_rmse_val = np.empty([len(parametrs['max_depth']), len(parametrs['n_estimators']), len(parametrs['feature_subsample_size'])])
rf_rmse_train = np.empty([len(parametrs['max_depth']), len(parametrs['n_estimators']), len(parametrs['feature_subsample_size'])])
rf_learning_time = np.empty([len(parametrs['max_depth']), len(parametrs['n_estimators']), len(parametrs['feature_subsample_size'])])
for i, max_depth in enumerate(parametrs['max_depth']):
    for j, n_estimators in enumerate(parametrs['n_estimators']):
        for k, feature_subsample_size in enumerate(parametrs['feature_subsample_size']):
            print(i, j, k)
            start = time.time()
            model = RandomForestMSE(n_estimators=n_estimators, max_depth=max_depth, feature_subsample_size=feature_subsample_size)
            rf_rmse_val[i, j, k], rf_rmse_train[i, j, k] = model.fit(X_train, y_train, X_val=X_test, y_val=y_test)
            finish = time.time()
            rf_learning_time[i, j, k] = finish - start
```

```python
print(rf_learning_time)
```

```python
parametrs = {'max_depth': [2, 4, 6, 8, 10, 12, None], 'n_estimators': [1, 20, 50, 100, 150, 200],
             'feature_subsample_size': [0.1, 0.3, 0.5, 0.7, 1], 'learning_rate': [0.1, 0.3, 0.6, 1]}

gb_rmse_val = np.empty([len(parametrs['max_depth']), len(parametrs['n_estimators']),
                    len(parametrs['feature_subsample_size']), len(parametrs['learning_rate'])])
gb_rmse_train = np.empty([len(parametrs['max_depth']), len(parametrs['n_estimators']),
                    len(parametrs['feature_subsample_size']), len(parametrs['learning_rate'])])
gb_learning_time = np.empty([len(parametrs['max_depth']), len(parametrs['n_estimators']),
                             len(parametrs['feature_subsample_size']), len(parametrs['learning_rate'])])
for i, max_depth in enumerate(parametrs['max_depth']):
    for j, n_estimators in enumerate(parametrs['n_estimators']):
        for k, feature_subsample_size in enumerate(parametrs['feature_subsample_size']):
            for l, learning_rate in enumerate(parametrs['learning_rate']):
                print(i, j, k, l)
                start = time.time()
                model = GradientBoostingMSE(n_estimators=n_estimators, max_depth=max_depth,
                                            feature_subsample_size=feature_subsample_size, learning_rate=learning_rate)
                gb_rmse_val[i, j, k, l], gb_rmse_train[i, j, k, l] = model.fit(X_train, y_train, X_val=X_test, y_val=y_test)
                finish = time.time()
                gb_learning_time[i, j, k, l] = finish - start
```

```python
np.save("gb_learning_time.npy", gb_learning_time)
np.save("rf_learning_time.npy", rf_learning_time)
np.save("gb_rmse_train.npy", gb_rmse_train)
np.save("rf_rmse_train.npy", rf_rmse_train)
np.save("gb_rmse_val.npy", gb_rmse_val)
np.save("rf_rmse_val.npy", rf_rmse_val)


# array_rain_fall_npy = np.load("saved-rain-fall-binary.npy")

# print("NumPy array: \n", array_rain_fall_npy)
# print("Shape: ", array_rain_fall_npy.shape)
# print("Data Type: ", array_rain_fall_npy.dtype.name)
```
